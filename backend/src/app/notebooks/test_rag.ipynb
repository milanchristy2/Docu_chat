```python
from langchain_core.prompts import PromptTemplate
prompt_template=PromptTemplate.from_template(
"""
You are expert AI assistant that can has information about Startup Companies
Context:
{context}
Task:Use the context to answer the user's question
Guidelines:
1.If the answer is not in the context,state that you do not know.Do not use external knowledge
2.If the user's question is unclear,ask for clarification based on the context
3.Keep the answer professional,concise and accurate
User Question:
{query}
Answer: 
"""
)
```

```python
import os
from pydantic import SecretStr
from langchain_groq import ChatGroq
groq_api_key=os.environ["GROQ_API_KEY"]
llm=ChatGroq(
    model="groq/compound-mini",
    api_key=SecretStr(groq_api_key)
)
```

```python
from langchain_core.messages import HumanMessage
msg=llm.invoke([HumanMessage("hello is it working?")])
print(msg.content)
```

```python
def retrieve_docs(query,k=4):
    return vectorstore.similarity_search(query,k=k)
```

```python
from typing import List
def generate_answer(prompt:str)->str:
    """
    Generate an answer using the ChatGroq LLM.
    
    Args:
        prompt (str): The question or prompt to send to the model.
    
    Returns:
        str: The generated answer from the model.
    
    """
    messages:List[List[HumanMessage]]=[[HumanMessage(content=prompt)]]
    response=llm.generate(messages)
    return response.generations[0][0].text    
```

```python
def rag_chain(query:str)->str:
    docs=retrieve_docs(query,k=3)
    context="\n\n".join([doc.page_content for doc in docs])
    prompt=prompt_template.format(context=context,query=query)
    answer=generate_answer(prompt)
    return answer
```

```python
query="what is bootlabs technologies do?"
ans=rag_chain(query)
print(ans)
```

```python
query="where is bootlabs situated?"
ans=rag_chain(query)
print(ans)
```
